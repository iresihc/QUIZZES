```python
##Creamos un diccionario con valores de diferentes tipos, string, valores numericos y flotantes
diccionario={"Amazon":[9.5,1.76,20,1,55],
             "Apple":[9.2,1.60,21,2,34],
             "Tesla":[9.7,1.77,28,3,45]}
```

```python
##Con esto mostramos los items que estan dentro de nuestro diccionario
diccionario.items()
```

```python
#Creamos un panel de datos con base a nuestro diccionario
import pandas as pd
df=pd.DataFrame(diccionario, index=["Interes", "Margen", "Credito","Deuda","Ventas"])
df
```

```python
##A partir de esta funcion, eliminamos columnas y es la primera segregacion
df.iloc[:,[0,2]]
```

```python
##Segunda segregacion, la cual elimina renglones
df.iloc[3:5]
```

```python
##Tercera segregacion que elimina dos columnas
df.drop(columns=["Amazon","Tesla"])
```

```python
df.describe()
#En la tabla se obtuvieron valores estadisticos como lo son la media(promedio), desviacion estandar y los cuartiles de los datos, ademas del #de datos que existen, los mayores y los menores
```

```python
##Este es el primer filtro de busqueda para ver los mayores valores dentro del diccionario
mayores= df[df["Amazon"]>10]
mayores= mayores.drop(columns=["Apple","Tesla"])
mayores
```

```python
##Este es el promedio de los datos
df.mean()
```

```python
##Este es la columna en la que se encuentra el minimo dato
min(df)
```
